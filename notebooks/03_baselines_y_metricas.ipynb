# 03 — Baselines y Métricas Comparativas

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve() / 'src'))

import pandas as pd
from discurso_odio.data.loader import load_dataset, prepare_dataset
from discurso_odio.features.preprocessing import preprocess_dataframe
from discurso_odio.models.baselines import get_baseline_models
from discurso_odio.models.evaluate import compare_models, evaluate_model, cross_validate_model

train = preprocess_dataframe(prepare_dataset(load_dataset('entrenamiento')))
test = preprocess_dataframe(prepare_dataset(load_dataset('pruebas')))
X_train, y_train = train['texto_limpio'], train['label']
X_test, y_test = test['texto_limpio'], test['label']

results = {}
for name, model in get_baseline_models().items():
    model.fit(X_train, y_train)
    ev = evaluate_model(model, X_test, y_test)
    cv = cross_validate_model(model, X_train, y_train, cv=3)
    results[name] = {**ev, 'cv': cv}

compare_models(results)
```
